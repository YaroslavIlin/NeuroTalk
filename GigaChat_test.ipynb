```python
from langchain_community.llms import GigaChat
from langchain.chains import LLMChain
from langchain_core.prompts import PromptTemplate
```

```python
# Авторизационные данные для работы с GigaChat API
authorization_data = "NzM2YjI0YzMtZWI3Zi00ODk4LWFhNjItNjZkNzUzZDFiNGFiOmU5YjZlYTNlLTYzMDctNGI2OC05N2NiLTlkNjY3OGFmOWIwYQ=="
```

```python
llm = GigaChat(credentials=authorization_data, scope="GIGACHAT_API_PERS", verify_ssl_certs=False)
```

```python
question_count = 2  # Количество вопросов от пользователя
for i in range(question_count):
  user_question = input("User: ")
  print()
  prompt = PromptTemplate.from_template(user_question)
  llm_chain = LLMChain(prompt=prompt, llm=llm)
  generated = llm_chain.invoke(input={})
  answer = generated["text"]
  print(f"GigaChat: {answer}")
  print()
  print("=" * 25)
  print()
```
